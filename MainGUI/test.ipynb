```python
from dataclasses import dataclass
from typing import List, Set

@dataclass
class Group:
    soma_indices: Set[int]
    rate: float
    distribution: str
```

```python
from collections import defaultdict

# Dictionary to hold tags and their respective groups
group_type = defaultdict(list)

# Example of adding groups to tags
group_type["sync"].append(Group(soma_indices={1, 2, 3}, rate=0.5, distribution='uniform'))
group_type["sync"].append(Group(soma_indices={4, 5, 6}, rate=0.8, distribution='normal'))
group_type["single"].append(Group(soma_indices={3, 4, 7}, rate=0.6, distribution='exponential'))
group_type["output"].append(Group(soma_indices={2, 8, 9}, rate=0, distribution='none'))
```

```python
for group in group_type["sync"]:
    print(f"Somas: {group.soma_indices}, Rate: {group.rate}, Distribution: {group.distribution}")
```
